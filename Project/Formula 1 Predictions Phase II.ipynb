```python
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
def formula1archivedata():
    dsdict1 = {} #rank
    dsdict2 = {} #team
    dsdict3 = {} #points
    for i in range(1950,2021):
        with open(f"WebData/{i}.html") as f:
            soup = BeautifulSoup(f, "html.parser")
        alist = soup.find_all("tr")
        clist = []
        for item in alist:
            blist = item.text
            clist.append(blist)
        for itema in clist[1:]:
            item = itema.split("\n")
            name = " ".join(item[5:7])
            if name not in dsdict1:
                dsdict1[name] = {}
                dsdict2[name] = {}
                dsdict3[name] = {}
                dsdict1[name][str(i)] = int(item[2])
                dsdict2[name][str(i)] = item[12]
                dsdict3[name][str(i)] = float(item[14])
            else:
                dsdict1[name][str(i)] = int(item[2])
                dsdict2[name][str(i)] = item[12]
                dsdict3[name][str(i)] = float(item[14])
    rank = pd.DataFrame.from_dict(dsdict1, orient="index")
    rank.fillna(value="None", inplace = True)
    team = pd.DataFrame.from_dict(dsdict2, orient="index")
    team.fillna(value="None", inplace = True)
    points = pd.DataFrame.from_dict(dsdict3, orient="index")
    points.fillna(value="None", inplace = True)
    rank.to_csv('Rank_by_Year.csv', index = True)
    team.to_csv('Team_by_Year.csv', index = True)
    points.to_csv('Points_by_Year.csv', index = True)
    pprint(rank)
    pprint(team)
    pprint(points)
    return "Complete"    
############ Function Call ############
formula1archivedata()
```

## Web Collection Requirement \#1


```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
def ergastdriverinfo():
    url = "http://ergast.com/api/f1/drivers?limit=853"
    request = requests.get(url)
    content = request.text
    soup = BeautifulSoup(content, "xml")
    GivenName = soup.find_all("GivenName")
    FamilyName = soup.find_all("FamilyName")
    DOB = soup.find_all("DateOfBirth")
    Nationality = soup.find_all("Nationality")
    didict = {}
    for i in range(0,len(GivenName)):
        name = f"{GivenName[i].get_text()} {FamilyName[i].get_text()}"
        didict[name] = {}
        didict[name]["Birthday"] = DOB[i].get_text()
        didict[name]["Nationality"] = Nationality[i].get_text()
    driverinfo = pd.DataFrame.from_dict(didict, orient="index")
    driverinfo.to_csv('Driver_Info.csv', index=True)
    pprint(driverinfo)
    return "Complete"
############ Function Call ############
ergastdriverinfo()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
import re
def ergasttrackinfo():
    url = "http://ergast.com/api/f1/circuits?limit=77"
    request = requests.get(url)
    content = request.text
    soup = BeautifulSoup(content, "xml")
    alist = soup.find_all("Circuit")
    circuitdict = {}
    for i in alist:
        course = i.find("CircuitName").get_text()
        city = i.find("Locality").get_text()
        country = i.find("Country").get_text()
        url1 = i.get('url')
        circuitdict[course] = {}
        circuitdict[course]["City"] = city
        circuitdict[course]["Country"] = country
        circuitdict[course]["URL"] = url1
    circuits = pd.DataFrame.from_dict(circuitdict, orient="index")
    circuits.to_csv("Circuits_Info.csv", index=True)
    pprint(circuits)
    return "Complete"
############ Function Call ############
ergasttrackinfo()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
def ergastracewinners():
    url1 = "http://ergast.com/api/f1/results/1?limit=1000"
    request1 = requests.get(url1)
    url2 = "http://ergast.com/api/f1/results/1?limit=1000&offset=1000"
    requests2 = requests.get(url2)
    content1 = request1.text
    content2 = requests2.text
    soup1 = BeautifulSoup(content1, "xml")
    soup2 = BeautifulSoup(content2, "xml")
    racename = soup1.find_all("RaceName") + soup2.find_all("RaceName") 
    circuit = soup1.find_all("CircuitName") + soup2.find_all("CircuitName")
    date = soup1.find_all("Date") + soup2.find_all("Date")
    winnergn = soup1.find_all("GivenName") + soup2.find_all("GivenName")
    winnerfn = soup1.find_all("FamilyName") + soup2.find_all("FamilyName")
    constructor = soup1.find_all("Name") + soup2.find_all("Name")
    racedict = {}
    for i in range(0, len(racename)):
        race = racename[i].get_text()
        year = date[i].get_text()[0:4]
        raceitem = f"{year} {race}"
        fn = winnergn[i].get_text()
        ln = winnerfn[i].get_text()
        winner = f"{fn} {ln}"
        racedict[raceitem] = {}
        racedict[raceitem]["Name"] = race
        racedict[raceitem]["Year"] = year
        racedict[raceitem]["Circuit"] = circuit[i].get_text()
        racedict[raceitem]["Winner"] = winner
        racedict[raceitem]["Constructor"] = constructor[i].get_text()
    del racedict["1951 French Grand Prix"]
    racevictor = pd.DataFrame.from_dict(racedict, orient="Index")
    racevictor.to_csv('Race_Winners.csv', index=True)
    pprint(racevictor)
    return racedict
        
        


############ Function Call ############
ergastracewinners()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
def ergastpolepos():
    url = "http://ergast.com/api/f1/grid/1/qualifying?limit=10000"
    request = requests.get(url)
    content = request.text
    soup = BeautifulSoup(content, 'xml')
    alist = soup.find_all("Race")
    polposdict = {}
    for i in alist:
        fn = i.find("GivenName").get_text()
        ln = i.find("FamilyName").get_text()
        name = f"{fn} {ln}"
        ssn = i.get("season")
        nm = i.find("RaceName").get_text()
        race = f"{ssn} {nm}"
        polposdict[race] = {}
        polposdict[race]["Pole Position"]=name
    polpos = pd.DataFrame.from_dict(polposdict, orient="index")
    polpos.to_csv("Pole_Position.csv", index=True)
    print(polpos)
    return polposdict
    
############ Function Call ############
ergastpolepos()
```

```python
def ergastpolewin():
    polposdict = ergastpolepos()
    racedict = ergastracewinners()
    for item in polposdict:
        polposdict[item]["Winner"]=racedict[item]["Winner"]
    pprint(polposdict)
    return "Complete"
    
############ Function Call ############
ergastpolewin()
```

```python
import re
import requests
import json
from bs4 import BeautifulSoup
from pprint import pprint
def statsf1worldchamp():
    url = "https://www.statsf1.com/en/statistiques/pilote/champion/nombre.aspx"
    with requests.Session() as s:
        s.headers = {
            "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.115 Safari/537.36",
            "Content-Type": "application/x-www-form-urlencoded",
            "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
            "Referer": "https://apps.neb-one.gc.ca/CommodityStatistics/Statistics.aspx",
            "Accept-Encoding": "gzip, deflate, br",
            "Accept-Language": "en-US,en;q=0.9"
        }
        response = s.get(url)
    soup = BeautifulSoup(response.content, 'html5lib')
    item = soup.find_all("tbody")[1]
    elements = item.find_all("tr")
    wcdict = {}
    for item in elements:
        data = item.find_all("td")
        rank = data[0].get_text()
        namestr = data[1].get_text()
        name = namestr.split()
        name.reverse()
        name = " ".join(name)
        yearstr = data[2].get_text()
        years = re.findall(r"[1-2][0-9]{3}", yearstr)
        num = data[4].get_text()
        wcdict[rank] = {}
        wcdict[rank]["Name"] = name.lower()
        wcdict[rank]["Years"] = years
        wcdict[rank]["Total Championships"] = num
    with open("World_Championships.json", "w") as f:
        json.dump(wcdict, f)
    pprint(wcdict)
    return "Complete"
############ Function Call ############
statsf1worldchamp()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
def statsf1grandslam():
    url = "https://www.statsf1.com/en/statistiques/pilote/divers/chelem.aspx"
    with requests.Session() as s:
        s.headers = {
            "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.115 Safari/537.36",
            "Content-Type": "application/x-www-form-urlencoded",
            "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
            "Referer": "https://apps.neb-one.gc.ca/CommodityStatistics/Statistics.aspx",
            "Accept-Encoding": "gzip, deflate, br",
            "Accept-Language": "en-US,en;q=0.9"
        }
        response = s.get(url)
    soup = BeautifulSoup(response.content, 'html5lib')
    item = soup.find_all("tbody")[2]
    elements = item.find_all("tr")
    gsdict = {}
    for item in elements:
        data = item.find_all("td")
        rank = data[0].get_text()
        namestr = data[1].get_text()
        name = namestr.split()
        name.reverse()
        name = " ".join(name)
        number = data[2].get_text()
        gsdict[rank] = {}
        gsdict[rank]["Name"] = name.lower()
        gsdict[rank]["Total"] = number
    with open("Grand_Slam.json", "w") as f:
        json.dump(gsdict, f)
    pprint(gsdict)
    return "Complete"
############ Function Call ############
statsf1grandslam()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
def driverdbscrape():
    url = "https://www.driverdb.com/championships/statistics/formula-1/"
    request = requests.get(url)
    content = request.text
    soup = BeautifulSoup(content, "html.parser")
    tables = soup.find_all("table")
    alist = ["Wins", "Podiums", "Fastest Laps", "Pole Positions", "Race Starts", "Race win percentage"]
    count = 0
    for i in tables:
        elements = i.find_all("tr")
        adict = {}
        for item in elements:
            data = item.find_all("td")
            position = data[0].get_text()
            name = data[2].get_text()
            stat = data[3].get_text()
            adict[position] = {}
            adict[position]["Name"] = name
            adict[position][f"{alist[count]}"] = stat
        dataframe = pd.DataFrame.from_dict(adict, orient="index")
        pprint(dataframe)
        dataframe.to_csv(f"All_Time_{alist[count]}.csv", index=True)
        count +=1
    return "Complete"   
############ Function Call ############
driverdbscrape()
```
